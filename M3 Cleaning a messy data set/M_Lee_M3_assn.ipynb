# 3. Data Preparation

## Step 1: Handling Missing Values
Missing values can be handled in several ways depending on the context:

* Deletion: If a column has a high percentage of missing values, it might be more informative to remove it entirely.
* Imputation: For columns with a small percentage of missing values, I can impute missing entries with a central tendency measure (mean, median) or use more complex methods like KNN imputation.

```python
from sklearn.impute import SimpleImputer, KNNImputer

# Initialize imputers
mean_imputer = SimpleImputer(strategy='mean')
median_imputer = SimpleImputer(strategy='median')
knn_imputer = KNNImputer(n_neighbors=5)

# Choosing imputation strategy based on data distribution and percentage of missing data
# Assume we've determined 'Alcohol' is normally distributed and has less missing data
wine_df['Alcohol'] = mean_imputer.fit_transform(wine_df[['Alcohol']])

# Assume 'ResidualSugar', 'Chlorides', 'FreeSulfurDioxide', 'TotalSulfurDioxide' are skewed or have outliers
for col in ['ResidualSugar', 'Chlorides', 'FreeSulfurDioxide', 'TotalSulfurDioxide']:
    wine_df[col] = median_imputer.fit_transform(wine_df[[col]])

# If 'pH' and 'Sulphates' are neither normally distributed nor skewed, we might opt for KNN imputation
for col in ['pH', 'Sulphates']:
    wine_df[col] = knn_imputer.fit_transform(wine_df[[col]])

# If 'LabelAppeal' is categorical or ordinal with a clear mode, we use the mode for imputation
mode_imputer = SimpleImputer(strategy='most_frequent')
wine_df['LabelAppeal'] = mode_imputer.fit_transform(wine_df[['LabelAppeal']])

# Deleting columns with too many missing values
# Here, we need to make a decision based on the percentage of missing values
# If a column like 'STARS' has a high percentage of missing values, we might opt to drop it
threshold = 0.5  # for example, if more than 50% data is missing, we drop the column
missing_percentages = wine_df.isnull().mean()
columns_to_drop = missing_percentages[missing_percentages > threshold].index
wine_df.drop(columns=columns_to_drop, inplace=True)
```

## Step 2: Correcting Data Errors
Data errors can be fixed by identifying out-of-range values or impossible values and either correcting them if possible or removing the affected records.

```python
# Removing rows where pH is outside the expected range for wine
wine_df = wine_df[(wine_df['pH'] >= 2.0) & (wine_df['pH'] <= 4.0)]
```

## Step 3: Handling Outliers
Outliers can distort the performance of machine learning models. They can be handled by:
* Capping: Using a predetermined threshold, such as the 1st and 99th percentiles.
* Transformation: Applying mathematical transformations to reduce the effect of extreme values.

```python
# Capping outliers at the 1st and 99th percentiles
for column in wine_df.columns:
    lower_limit = wine_df[column].quantile(0.01)
    upper_limit = wine_df[column].quantile(0.99)
    wine_df[column] = np.clip(wine_df[column], lower_limit, upper_limit)
```

## Step 4: Feature Engineering
New variables can be created to enhance the dataset's predictive power:
* Ratios: Creating new features that are ratios of existing features.
* Binning: Transforming continuous variables into categorical bins.

```python
# Creating a new feature as a ratio of FreeSulfurDioxide to TotalSulfurDioxide
wine_df['SulfurRatio'] = wine_df['FreeSulfurDioxide'] / wine_df['TotalSulfurDioxide']

# Binning Alcohol content into categories
wine_df['AlcoholBinned'] = pd.cut(wine_df['Alcohol'], bins=[0, 5, 10, 15, 20, np.inf], labels=['Low', 'Medium', 'High', 'Very High', 'Infinite'])
```

## Step 5: Encoding Categorical Variables
Machine learning algorithms usually require numerical input, so categorical variables need to be converted into a numerical format.

```python
# Encoding the 'STARS' categorical variable using one-hot encoding
stars_dummies = pd.get_dummies(wine_df['STARS'], prefix='STAR')
wine_df = pd.concat([wine_df, stars_dummies], axis=1)
wine_df.drop('STARS', axis=1, inplace=True)
```

## Step 6: Data Scaling
Many machine learning algorithms perform better when numerical input variables are on the same scale.

```python
from sklearn.preprocessing import StandardScaler

# Assuming wine_df is your original DataFrame

# Select only the numeric columns for scaling
numeric_cols = wine_df.select_dtypes(include=[np.number]).columns

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the scaler on the numeric columns
wine_df_scaled = pd.DataFrame(scaler.fit_transform(wine_df[numeric_cols]), columns=numeric_cols)

# Now you have a DataFrame with scaled numeric values
# If you want to include the non-numeric columns, you can concatenate them back
non_numeric_cols = wine_df.select_dtypes(exclude=[np.number]).columns
wine_df_scaled = pd.concat([wine_df_scaled, wine_df[non_numeric_cols].reset_index(drop=True)], axis=1)
```

## Step 7: Data Transformation
Transformations can help in stabilizing variance and making the data more normal distribution-like, which can improve the performance of some algorithms.

```python
# Applying a log transformation to skewed features
skewed_features = ['ResidualSugar', 'Chlorides']
for feature in skewed_features:
    wine_df[feature] = np.log1p(wine_df[feature])
```
